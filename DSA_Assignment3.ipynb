# 1. Number of males and females participated

# 2. Students' parental level education

```python
data['parental level of education'].value_counts(normalize = True)
```

```python
# Most of the parents level of education is from some college
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Parents education vs scores for all three subjects:', fontsize=30)


sns.boxplot(x='parental level of education', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Parents education')
var1 = ax1.set_xticklabels(ax1.get_xticklabels(), rotation=30)
 
    
sns.boxplot(x='parental level of education', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Parents education')
var2 = ax2.set_xticklabels(ax2.get_xticklabels(), rotation=30)


sns.boxplot(x='parental level of education', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Parents education')
var3 = ax3.set_xticklabels(ax3.get_xticklabels(), rotation=30)
```

```python
# So, the students who have scored the highest in all three subjects belong to the parents whose education level is 
# master's degree.

# And also the students who have scored lowest in all three subjects belong to the parents whose education level is high school 
```

# 3. Who scores the most?

```python
# Based on gender
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20) 
# Math score
sns.boxplot(x='gender', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Gender')
# Reading Score 
sns.boxplot(x='gender', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Gender')
# Writing Score
sns.boxplot(x='gender', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Gender')
```

```python
# On average, male students has scored the most than females
# On the other hand in both reading and writing female students scored the most
```

```python
# Based on test preparation course
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20) 
# Math score
sns.boxplot(x='test preparation course', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Test preparation course')
# Reading Score 
sns.boxplot(x='test preparation course', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Test preparation course')
# Writing Score
sns.boxplot(x='test preparation course', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Test preparation course')
```

```python
# The students who has completed the test preparation course has scored the most on average.
```

# 4. Score variation

```python
# Based on gender
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20) 
# Math score
sns.boxplot(x='gender', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Gender')
# Reading Score 
sns.boxplot(x='gender', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Gender')
# Writing Score
sns.boxplot(x='gender', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Gender')
```

```python
# From the above plots it is clear that male students scored the highest on an average. There is student who scored zero in maths.
# who is a female student.

# In reading, female students scored the most. But also, a  female student got the lowest mark.

# Also in writing, female students score the most and the lowest mark was also from a female student.
```

```python
# Based on test preparation course
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20) 
# Math score
sns.boxplot(x='test preparation course', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Test preparation course')
# Reading Score 
sns.boxplot(x='test preparation course', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Test preparation course')
# Writing Score
sns.boxplot(x='test preparation course', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Test preparation course')
```

```python
# All the students who have completed the test preparation course scored better marks than the other students who did not 
# complete it.
# Also the students who got lower marks are those students who didn't complete the test preparation course.
```

# 5. Bonus points to the top 25% students

```python
l = data["math score"].quantile(0.75)
top25 = data[data["math score"] >= l].sort_values('math score',ascending=False)
top = top25.head(250)
top['bonus points'] = 20
top
```
